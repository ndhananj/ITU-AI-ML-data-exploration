```python
import pandas as pd
import os
```

```python
df = pd.read_csv('Train_ULAK.csv')
df
```

```python
import numpy as np
```

```python
y=df[[' Label']]
np.unique(y)
```

```python
x=df.drop([' Label'], axis=1)
x
```

```python
x_na = x[x.isnull().any(axis=1)]
x_na
```

```python
x_nna = x.dropna()
x_nna
```

```python
[(c,x.loc[x[c].idxmax()][c]) for c in x.columns]
```

```python
x_inf=x[np.isinf(x).any(axis=1)]
x_nna_inf=x_nna[np.isinf(x_nna).any(axis=1)]

x_clean = x_nna[~np.isinf(x_nna).any(axis=1)]
x_clean
```

```python
[(c,x.loc[x_clean[c].idxmax()][c]) for c in x.columns]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(x_clean.values)
```

```python
print(pca.explained_variance_ratio_)
```

```python
gb=df.groupby(' Label')
dfs = [gb.get_group(x) for x in gb.groups]
xs = [ f.drop([' Label'], axis=1) for f in dfs]
x_cleans = [x[~np.isinf(x).any(axis=1)].dropna() for x in xs] 
x_pcas = [pca.transform(x_clean.values) for x_clean in x_cleans] 
x_pcas
```
